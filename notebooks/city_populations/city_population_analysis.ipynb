```python
# Phân tích dữ liệu dân số và mật độ thành phố

## Cài đặt các thư viện cần thiết
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

# Cấu hình biểu đồ đẹp hơn
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 12
sns.set_palette("viridis")

# Bỏ qua các cảnh báo
import warnings
warnings.filterwarnings('ignore')

## Đọc dữ liệu
# Đọc dữ liệu từ file CSV
df = pd.read_csv('./datasets/city_populations/city_population_density.csv')

# Hiển thị thông tin cơ bản về bộ dữ liệu
print(f"Kích thước dữ liệu: {df.shape}")
print("\nThông tin cơ bản về các cột:")
display(df.info())

print("\n5 dòng đầu tiên của dữ liệu:")
display(df.head())

## Tiền xử lý dữ liệu
# Kiểm tra các giá trị thiếu
print("\nKiểm tra giá trị thiếu:")
display(df.isnull().sum())

# Chuyển đổi các cột số sang định dạng số
# Loại bỏ các ký tự không phải số từ các cột dân số, diện tích và mật độ
df['Population'] = pd.to_numeric(df['Population'].str.replace(',', ''), errors='coerce')
df['Area KM2'] = pd.to_numeric(df['Area KM2'].str.replace(',', ''), errors='coerce')
df['Area   M2'] = pd.to_numeric(df['Area   M2'].str.replace(',', ''), errors='coerce')
df['Density KM2'] = pd.to_numeric(df['Density KM2'].str.replace(',', ''), errors='coerce')
df['Density  M2'] = pd.to_numeric(df['Density  M2'].str.replace(',', ''), errors='coerce')

# Hiển thị thông tin thống kê cơ bản sau khi xử lý
print("\nThống kê mô tả sau khi xử lý:")
display(df.describe())

```

```python
## Task 1: Sắp xếp và in ra 10 thành phố có dân số lớn nhất & 10 thành phố có dân số nhỏ nhất
print("\n## Task 1: Top 10 thành phố có dân số lớn nhất")
top_10_pop = df.sort_values('Population', ascending=False).head(10)
display(top_10_pop[['City', 'Population', 'Country']])

print("\n## Task 1: Top 10 thành phố có dân số nhỏ nhất")
bottom_10_pop = df.sort_values('Population').head(10)
display(bottom_10_pop[['City', 'Population', 'Country']])

# Vẽ biểu đồ top 10 thành phố có dân số lớn nhất
plt.figure(figsize=(14, 8))
sns.barplot(x='Population', y='City', data=top_10_pop, palette='viridis')
plt.title('Top 10 thành phố có dân số lớn nhất')
plt.xlabel('Dân số')
plt.ylabel('Thành phố')
plt.ticklabel_format(style='plain', axis='x')
plt.tight_layout()
plt.show()

# Vẽ biểu đồ top 10 thành phố có dân số nhỏ nhất
plt.figure(figsize=(14, 8))
sns.barplot(x='Population', y='City', data=bottom_10_pop, palette='viridis')
plt.title('Top 10 thành phố có dân số nhỏ nhất')
plt.xlabel('Dân số')
plt.ylabel('Thành phố')
plt.ticklabel_format(style='plain', axis='x')
plt.tight_layout()
plt.show()

```

```python
## Task 2: In ra tên các quốc gia có tối thiểu 3 thành phố trong danh sách này
country_counts = df['Country'].value_counts()
countries_min_3_cities = country_counts[country_counts >= 3]

print("\n## Task 2: Quốc gia có tối thiểu 3 thành phố trong danh sách")
display(countries_min_3_cities)

# Vẽ biểu đồ quốc gia có tối thiểu 3 thành phố
plt.figure(figsize=(14, 8))
sns.barplot(x=countries_min_3_cities.values, y=countries_min_3_cities.index, palette='viridis')
plt.title('Quốc gia có tối thiểu 3 thành phố trong danh sách')
plt.xlabel('Số lượng thành phố')
plt.ylabel('Quốc gia')
plt.tight_layout()
plt.show()

```

```python
## Task 3: In ra Top 5 quốc gia có nhiều thành phố xuất hiện trong bảng này nhất
top_5_countries = country_counts.sort_values(ascending=False).head(5)

print("\n## Task 3: Top 5 quốc gia có nhiều thành phố xuất hiện nhất")
display(top_5_countries)

# Vẽ biểu đồ top 5 quốc gia có nhiều thành phố nhất
plt.figure(figsize=(12, 6))
sns.barplot(x=top_5_countries.values, y=top_5_countries.index, palette='viridis')
plt.title('Top 5 quốc gia có nhiều thành phố xuất hiện nhất')
plt.xlabel('Số lượng thành phố')
plt.ylabel('Quốc gia')
plt.tight_layout()
plt.show()

```

```python
## Task 4: In ra các thành phố có dân số & diện tích đều nằm trong Top 20
# Xác định top 20 thành phố có dân số lớn nhất
top_20_population = set(df.nlargest(20, 'Population')['City'])

# Xác định top 20 thành phố có diện tích lớn nhất
top_20_area = set(df.nlargest(20, 'Area KM2')['City'])

# Tìm giao của hai tập hợp
cities_in_both_top_20 = top_20_population.intersection(top_20_area)

print("\n## Task 4: Thành phố có dân số & diện tích đều nằm trong Top 20")
cities_in_both = df[df['City'].isin(cities_in_both_top_20)]
display(cities_in_both[['City', 'Population', 'Area KM2', 'Country']])

# Tạo một biểu đồ scatter plot cho dân số và diện tích
plt.figure(figsize=(14, 8))
sns.scatterplot(x='Area KM2', y='Population', data=df, alpha=0.6, s=100)

# Đánh dấu các thành phố nằm trong cả hai top 20
for idx, row in cities_in_both.iterrows():
    plt.annotate(row['City'], (row['Area KM2'], row['Population']), 
                 xytext=(5, 5), textcoords='offset points')
    plt.scatter(row['Area KM2'], row['Population'], color='red', s=120, zorder=5)

plt.title('Mối quan hệ giữa diện tích và dân số của các thành phố')
plt.xlabel('Diện tích (km²)')
plt.ylabel('Dân số')
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
## Task 5: Thống kê mật độ dân số theo quốc gia
# Tính mật độ dân số trung bình, lớn nhất và nhỏ nhất theo quốc gia
country_density_stats = df.groupby('Country')['Density KM2'].agg(['mean', 'min', 'max', 'count']).sort_values('mean', ascending=False)

print("\n## Task 5: Thống kê mật độ dân số theo quốc gia")
display(country_density_stats)

# Lọc các quốc gia có ít nhất 2 thành phố để biểu đồ rõ ràng hơn
countries_with_multiple_cities = country_density_stats[country_density_stats['count'] >= 2]

# Vẽ biểu đồ mật độ dân số trung bình theo quốc gia
plt.figure(figsize=(14, 8))
sns.barplot(x=countries_with_multiple_cities['mean'], y=countries_with_multiple_cities.index, palette='viridis')
plt.title('Mật độ dân số trung bình theo quốc gia (km²)')
plt.xlabel('Mật độ dân số trung bình')
plt.ylabel('Quốc gia')
plt.ticklabel_format(style='plain', axis='x')
plt.tight_layout()
plt.show()

# Vẽ biểu đồ boxplot mật độ dân số theo quốc gia
plt.figure(figsize=(14, 8))
# Chỉ lọc các quốc gia có từ 2 thành phố trở lên để boxplot có ý nghĩa
countries_for_boxplot = country_counts[country_counts >= 2].index
df_filtered = df[df['Country'].isin(countries_for_boxplot)]
sns.boxplot(x='Density KM2', y='Country', data=df_filtered, palette='viridis')
plt.title('Phân bố mật độ dân số theo quốc gia (km²)')
plt.xlabel('Mật độ dân số')
plt.ylabel('Quốc gia')
plt.ticklabel_format(style='plain', axis='x')
plt.tight_layout()
plt.show()

```

```python
## Task 6: Thống kê các thành phố có dân số lớn nhất của từng quốc gia
# Chỉ xem xét các quốc gia có ít nhất 2 thành phố
countries_with_multiple_cities = country_counts[country_counts >= 2].index

# Tìm thành phố có dân số lớn nhất của mỗi quốc gia
largest_cities_by_country = {}
for country in countries_with_multiple_cities:
    country_data = df[df['Country'] == country]
    largest_city = country_data.loc[country_data['Population'].idxmax()]
    largest_cities_by_country[country] = {
        'City': largest_city['City'],
        'Population': largest_city['Population'],
        'Area KM2': largest_city['Area KM2'],
        'Density KM2': largest_city['Density KM2']
    }

# Chuyển đổi thành DataFrame để hiển thị
largest_cities_df = pd.DataFrame.from_dict(largest_cities_by_country, orient='index')
largest_cities_df = largest_cities_df.sort_values('Population', ascending=False)

print("\n## Task 6: Thành phố có dân số lớn nhất của từng quốc gia (có ít nhất 2 thành phố)")
display(largest_cities_df)

# Vẽ biểu đồ thành phố có dân số lớn nhất của từng quốc gia
plt.figure(figsize=(14, 8))
plt.bar(largest_cities_df.index, largest_cities_df['Population'], color=sns.color_palette("viridis", len(largest_cities_df)))
plt.xticks(rotation=45, ha='right')
plt.title('Dân số của thành phố lớn nhất theo quốc gia')
plt.xlabel('Quốc gia')
plt.ylabel('Dân số')
plt.ticklabel_format(style='plain', axis='y')
plt.tight_layout()
plt.show()

# Vẽ biểu đồ kết hợp giữa dân số và mật độ
plt.figure(figsize=(14, 8))
plt.scatter(largest_cities_df['Population'], largest_cities_df['Density KM2'], s=largest_cities_df['Area KM2']/100, alpha=0.7)

# Ghi chú tên thành phố
for idx, row in largest_cities_df.iterrows():
    plt.annotate(f"{row['City']} ({idx})", 
                 (row['Population'], row['Density KM2']),
                 xytext=(5, 5), textcoords='offset points')

plt.title('Mối quan hệ giữa dân số và mật độ dân số của thành phố lớn nhất mỗi quốc gia')
plt.xlabel('Dân số')
plt.ylabel('Mật độ dân số (km²)')
plt.ticklabel_format(style='plain')
plt.grid(True)
plt.tight_layout()
plt.show()

```
